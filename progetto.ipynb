# Pulizia dei dati

Si prende la colonna contenente i luoghi "Initial Solution" (che Ã¨ una serie di pandas) e tramite la funzione str.replace() si pulisce la stringa, rimuovendo la punteggiatura speciale (parentesi quadre, virgole e apici) con un carattere null.

Si stampa poi la colonna "Initial Solution" ripulita.

```python
data["Initial Solution"] = data["Initial Solution"].str.replace("[","")
data["Initial Solution"] = data["Initial Solution"].str.replace("]","")
data["Initial Solution"] = data["Initial Solution"].str.replace(",","")
data["Initial Solution"] = data["Initial Solution"].str.replace("'","")

data["Initial Solution"].head()
```

# Divisione dei luoghi

Si "isola" la singola colonna "Initial Solution" inserendola nella variabile "init_sol", che diventa una serie a sua volta.

```python
init_sol = data["Initial Solution"]
print(init_sol)
type(init_sol)
```

La singola riga della serie equivale ad una stringa. Ad esempio "init_sol[0]" equivale ad una stringa contenente "D0 C21 C26 D0 D0 C65 C61 D0 D0 C24 C50 D0 ..."

Attraverso il metodo split(), si prende una riga della serie "init_sol" (in questo caso la riga 0) e la si splitta rispetto al carattere spazio. Ogni volta che il metodo split() trova uno spazio vuoto, viene rilevato un elemento (luogo).

La stringa splittata per singolo elemento viene inserita in una lista chiamata "places", dove sono presenti ancora ripetizioni.

```python
places = init_sol[0].split()
#print(" ".join(sorted(set(places), key=places.index)))

print("places Ã¨ una:", type(places))
print("\nL'intera lista 'places' Ã¨:\n", places)
print("\nIn prima posizione c'Ã¨:", places[0])
print("C'Ã¨ ridondanza. Al quarto posto della lista abbiamo lo stesso elemento:", places[3])
```

Si istanzia una nuova lista che conterrÃ  i singoli luoghi senza ripetizioni, chiamata "single_places".

Attraverso un ciclo for vengono iterati tutti gli elementi nella lista originaria "places" e, se non presenti nella nuova lista "single_places", vengono inseriti tramite append() in quest'ultima lista.

Si stampa poi la lista senza ripetizioni.

```python
single_places = []
for i in places:
    if i not in single_places:
        single_places.append(i)
        
print("Lista dei luoghi senza ridondanza:\n", single_places)
print("\nPrint di diversi luoghi:\n", single_places[0], single_places[5], single_places[23])
```
